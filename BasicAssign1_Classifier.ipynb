```python
#Basic Assignment 1, Classifier

#Import required packages

#Import Scikit-Learn
import sklearn

#For this question, you will need to import the decision tree 
#and kappa classes
from sklearn import tree
from sklearn.metrics import cohen_kappa_score

#Use the following classifier and hyperparameters:
#tree.DecisionTreeClassifier(min_samples_split=10)

# Import other packages you will need
import numpy as np
import pandas as pd
import os
#import csv

#Find where this notebook is in case you need a filepath 
#to load the dataset
print(os.getcwd())
```

```python
#Read in the dataset using pandas
#If dataset is in the same folder with this file it should read it in.
data = pd.read_csv('Godwin_dataset.csv')

#Example of using the full file path just in case.
#data = pd.read_csv('C:/Users/Put/Your/Filepath/Here/Godwin_dataset.csv')

#Check the file
data.head()
```

```python
#It is important to note that Python’s native implementation of 
#decision trees uses the CART(Classification and Regression Trees) algorithm.

#CART can only input numerical variables, so you will have to transform 
#string-type variables to numerical variables.

#Your current data set has string-type variables.

#In order to allow CART to use these variables, 
#you will need to perform one-hot-encoding on all your string-type variables.

data1 = pd.get_dummies(data, columns = ['SCHOOL', 'Class', 'CODER ', 'Activity'])
###Note: there are some extra spaces in column names which is a common problem
###with real data.

#Take another look at the data
data1.head()
```

```python
#You will then need to create an array containing only the label,
#or the variable you want to predict.

#In this case, the label is ONTASK.
y = np.array(data[["ONTASK"]])

#Take a look at the array
y
```

```python
#You will then need to remove ONTASK 
#from your data set before building your decision tree.

X = data1.drop(columns=['ONTASK'])

#What does the file look like now?
X.head()
```

```python
#You are now ready to build your decision tree.

#Feed your set of features and the list of labels 
#into the DecisionTreeClassifier operator.
clf = tree.DecisionTreeClassifier(min_samples_split=10)
clf.fit(X, y)

#Can you compute the kappa now?
```

```python
#In order to compute kappa, you will first need to get
#the values predicted by your model.

predictions = clf.predict(X)

#What does this array look like?
predictions
```

```python
#Next, you will need to compare the predictions 
#to the actual labels in order to compute kappa.
kappa = cohen_kappa_score(y, predictions)

#Finally, print kappa.
print(kappa)

#The correct answer is 0.73
```

```python
#Question 7 of 10:
#What is the kappa, if you delete 
#School, Class, Coder, UNIQUEID, and STUDENTID,
#use CART, and conduct 10-fold student-level batch cross-validation? 
#Please use 2 decimal places.

#In order to perform batch cross-validation, 
#you will first need to import the necessary module: 
from sklearn.model_selection import GroupKFold
```

```python
#You will first need to ensure that each student appears 
#in ONLY ONE batch. 

#This way, for example, if the model is trained using Student A’s data,
#the model cannot also be trained [TESTED?] on Student A’s data

#You can do this by creating a dictionary of unique student IDs,
#each linked to a specific index number.

#I've changed "s_id" from assignment to "student_id"

group_dict = {}
groups = np.array([])
for index, row in data.iterrows():
    student_id = row['STUDENTID']
    if student_id not in group_dict:
        group_dict[student_id] = index
    groups = np.append(groups, group_dict[student_id])
```

```python
#You now need to create the batching variable.
#Remember, you are creating 10 folds 
#(i.e., each studentcan only be assigned to one of 10 groups).

gkf = GroupKFold(n_splits=10)

#You are now ready to build your model,
#this time using batch cross-validation, 
#iterating through your folds using the 
#batching variable you just created:

#Be sure to create a list of variables 
#outside the for-loop to store all 10 kappa values. 
kappaValues = []

for train_index, test_index in gkf.split(X, y, groups=groups):
    
#All other code stubs in the following hints 
#will be within the newly created for-loop, 
#unless otherwise stated.
#Using the training and testing indices, 
#pull the relevant features (from X) 
#and labels (from y) before building a model in this iteration:

    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y[train_index]
    y_test = y[test_index]
    
    clf = tree.DecisionTreeClassifier(min_samples_split=10)
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    
    kappa = cohen_kappa_score(y_test, predictions)
    kappaValues.append(kappa)

kappaValues
#kappaValues[1]

#Be sure to create a list of variables 
#outside the for-loop to store all 10 kappa values. 
```

```python
#Import mean
from statistics import mean

#Get the averagekappa and print
mean(kappaValues)

#Answer is 0.10
```

```python
#Question 9 of 10: What is the kappa, 
#for the same set of variables you used for question 4, 
#if you useGradient Boosting, 
#and conduct 10-fold student-level batch cross-validation?

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GroupKFold
```

```python
#Using the same batching and for-loop created in Question 7,
#replace the existing GaussianNB with the GradientBoostingClassifieroperator

#Repeating batching and for loop below for clarity

#You now need to create the batching variable.
#Remember, you are creating 10 folds 
#(i.e., each studentcan only be assigned to one of 10 groups).

gkf = GroupKFold(n_splits=10)

#You are now ready to build your model,
#this time using batch cross-validation, 
#iterating through your folds using the 
#batching variable you just created:

#Be sure to create a list of variables 
#outside the for-loop to store all 10 kappa values. 
kappaValuesGB = []

for train_index, test_index in gkf.split(X, y, groups=groups):
    
#All other code stubs in the following hints 
#will be within the newly created for-loop, 
#unless otherwise stated.
#Using the training and testing indices, 
#pull the relevant features (from X) 
#and labels (from y) before building a model in this iteration:

    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y[train_index]
    y_test = y[test_index]
    
    gb = GradientBoostingClassifier(learning_rate = .5\
                                     , n_estimators = 200\
                                     , min_samples_split = .1\
                                     , min_samples_leaf = .05\
                                     , min_weight_fraction_leaf = .05\
                                     , random_state=5)
    #flatten labels with GaussianNB
    gb.fit(X_train, y_train.ravel())
    predictions = gb.predict(X_test)
    
    kappa = cohen_kappa_score(y_test, predictions)
    kappaValuesGB.append(kappa)

kappaValuesGB
```
