```python
#Basic Assignment 3: Clustering

#Question 1: Conduct k-means clustering with K=2, using k-Means. 
#Look at the Cluster Centers. 
#Which attributes have the biggest difference between cluster 0 and cluster 1?

#First, you need to load your data into a dataframe. 
#You can do this using the pandas module. 
#In order to do this, you first need to import it

import pandas as pd

#Also import. . .
import sklearn
import os

#Read in the dataset as Pandas dataframe using pandas
#If dataset is in the same folder with this file it should read it in.
data = pd.read_csv('asgn7-clustering.csv')

#Example of using the full file path just in case.
#data = pd.read_csv('C:/Users/Put/Your/Filepath/Here/Godwin_dataset.csv')

#Check data
data.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from matplotlib import pyplot as plt

#fit data to agglomerative clustering model
#K = 2
aggc = AgglomerativeClustering(n_clusters=2).fit(X_1)
```

```python
def plot_dendrogram(model, **kwargs):
#Children of hierarchical clustering
    children = model.children_

#Distances between each pair of children
#Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

#The number of observations contained in eahc cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

#Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

#plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

#Finally, plot your dendrogram using the method defined in the previous hint
plt.show(plot_dendrogram(aggc, labels=aggc.labels_))
```
